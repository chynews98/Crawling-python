```
import requests
from bs4 import BeautifulSoup

num = 0

print('1 : 패션의류')
print('2 : 패션잡화')
print('3 : 화장품/미용')
print('4 : 디지털/가전')
print('5 : 가구/인테리어')
print('6 : 출산/육아')
print('7 : 스포츠/레저')
print('8 : 식품')
print('9 : 생활/건강')
print()

inp = input('어떤 카페고리의 순위를 조회하시겠습니까(숫자로, 네이버쇼핑 기준)? ')

inp = int(inp)

if inp == 1:
  num = 0
elif inp == 2:
  num = 1
elif inp == 3:
  num = 2
elif inp == 4:
  num = 3
elif inp == 5:
  num = 4
elif inp == 6:
  num = 5
elif inp == 7:
  num = 7
elif inp == 8:
  num = 6
elif inp == 9:
  num = 8

url = 'https://search.shopping.naver.com/best100v2/detail.nhn?catId=5000000' + str(num)

req = requests.get(url)

html = req.text
parse = BeautifulSoup(html, 'html.parser')

tem = parse.find_all('p', {'class' : 'cont'})

print('다음은 선택하신 카테고리의 인기상품 순위입니다.')
print()

for i in range(3):
  print(i + 1, '위 :', tem[i].text.strip())
```
