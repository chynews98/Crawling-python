```python
import requests
from bs4 import BeautifulSoup

url = 'http://dart.fss.or.kr/dsac001/mainY.do'
r = requests.get(url)
r = r.text
r = BeautifulSoup(r, 'html.parser')

r = r.find_all('span',{'class': 'nobr1'})
for i in r:
    url = 'http://dart.fss.or.kr/' + i.find('a')['href']
    r = requests.get(url)
    r = r.text
    r = BeautifulSoup(r, 'html.parser')
    print(r.text.strip())
    print()
```
